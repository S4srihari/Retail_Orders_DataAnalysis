```python
#import the dataset using Pandas
import pandas as pd
df = pd.read_csv(r'retail-orders\orders.csv')
df.head(10)
```

```python
#Identifying Wrong Format Data
```

```python
df['Ship Mode'].unique()
```

```python
#Handling wrong format Data
df.replace({'Ship Mode': ['Not Available', 'unknown']},'nan', inplace = True)
df['Ship Mode'].unique()
```

```python
#Identifying Inconsitencies in column naming
df.columns
```

```python
#Handling Inconsistency by making them consistent
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", '_')
df.columns
```

```python
df.head()
```

```python
#Found something is missing in data and adding a new column by calculating
df['discount_amount'] = df['list_price']*df['discount_percent']*0.01
df.head()
```

```python
#Finding Sell Price and adding data
df['sell_price'] = df['list_price'] - df['discount_amount']
```

```python
#Finding profit and adding in Data
df['profit'] = df['sell_price']-df['cost_price']
df.head()
```
